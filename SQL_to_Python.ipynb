```python
import sqlalchemy
from sqlalchemy import create_engine, func
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session

import pandas as pd

# NOT UPLOADED TO GITHUB
from config import PASSWORD, USERNAME, DATABASE_NAME, ENDPOINT
```

```python
engine = create_engine(f'postgresql+psycopg2://{USERNAME}:{PASSWORD}@{ENDPOINT}/{DATABASE_NAME}')
```

```python
Base = automap_base()
```

```python
Base.prepare(engine, reflect=True)
```

```python
# checking that tables have been connected
Base.classes.keys()
```

```python
#Assigning tables to variables
listings = Base.classes.listings
calendar = Base.classes.calendar
reviews = Base.classes.reviews
```

```python
session = Session(engine)
```

```python
# testing connection by pulling ids
listings_id = session.query(listings.id)
print(listings_id.all())
```

```python
# Querying everything from the listings table and converted to dataframe
sql_query_listings = pd.read_sql_query ('''
                               SELECT
                               *
                               FROM listings
                               ''', engine)


listings_df = pd.DataFrame(sql_query_listings, columns = sql_query_listings.keys())
listings_df.head(10)
```
